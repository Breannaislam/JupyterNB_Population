```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

datadir = '/content/drive/My Drive/Assignment 3/data/'

csv_file = datadir + 'population-and-demography.csv'
population_data = pd.read_csv(csv_file)

print(population_data.head())
```

```python
import plotly.express as px

def plot_population_interactive(frame, country):

    country_data = frame[frame['Country name'] == country]

    fig = px.line(
        country_data,
        x='Year',
        y='Population',
        title=country,
        labels={'Population': 'Population', 'Year': 'Year'},
        template='plotly'
    )
    fig.update_layout(title_font_size=20, title_x=0.5)
    fig.show()

plot_population_interactive(population_data, "United States")
```

# **Exercise 2**

```python
import plotly.express as px

def plot_populations(frame, countries):

    filtered_data = frame[frame['Country name'].isin(countries)].copy()

    filtered_data['Country name'] = pd.Categorical(
        filtered_data['Country name'], categories=sorted(countries), ordered=True
    )

    fig = px.line(
        filtered_data,
        x='Year',
        y='Population',
        color='Country name',
        title='Population',
        labels={'Population': 'Population', 'Year': 'Year', 'Country name': 'Countries'},
        template='plotly'
    )
    fig.update_layout(title_font_size=20, title_x=0.5)
    fig.show()

countries_to_plot = ['United States', 'Canada', 'Europe (UN)']
plot_populations(population_data, countries_to_plot)
```

## **Exercise 4**

```python
import plotly.express as px

def plot_by_age_group(frame, country):

    country_data = frame[frame['Country name'] == country]

    age_group_data = pd.DataFrame({
        'Year': country_data['Year'],
        'Pop under 15': country_data['Population of children under the age of 15'],
        'Pop between 15 and 64': country_data['Population aged 15 to 64 years'],
        'Pop over 64': country_data['Population'] -
                       (country_data['Population of children under the age of 15'] +
                        country_data['Population aged 15 to 64 years'])
    })

    melted_data = age_group_data.melt(id_vars='Year',
                                      var_name='Age Group',
                                      value_name='Population')

    fig = px.bar(
        melted_data,
        x='Year',
        y='Population',
        color='Age Group',
        title=country,
        labels={'Population': 'Population', 'Year': 'Year', 'Age Group': 'Age Groups'},
        template='plotly'
    )
    fig.update_layout(barmode='stack', title_font_size=20, title_x=0.5)
    fig.show()

plot_by_age_group(population_data, 'United States')
```

# **Exercise 5**

```python
import plotly.express as px

def plot_age_group_by_percentage(frame, country):

    country_data = frame[frame['Country name'] == country]

    country_data['Under 15'] = 100 * country_data['Population of children under the age of 15'] / country_data['Population']
    country_data['Between 15 and 64'] = 100 * country_data['Population aged 15 to 64 years'] / country_data['Population']
    country_data['Over 64'] = 100 - (country_data['Under 15'] + country_data['Between 15 and 64'])

    percentage_data = country_data.melt(
        id_vars='Year',
        value_vars=['Under 15', 'Between 15 and 64', 'Over 64'],
        var_name='Age Group',
        value_name='Percentage'
    )

    fig = px.bar(
        percentage_data,
        x='Year',
        y='Percentage',
        color='Age Group',
        title=country,
        labels={'Percentage': 'Population Percentage', 'Year': 'Year', 'Age Group': 'Age Groups'},
        template='plotly'
    )
    fig.update_layout(barmode='stack', title_font_size=20, title_x=0.5)
    fig.show()

plot_age_group_by_percentage(population_data, 'Japan')
```

```python
import plotly.express as px

def plot_population_map(frame, year):

    filtered_data = frame[frame['Year'] == year]

    fig = px.scatter_geo(
        filtered_data,
        locations="Country name",
        locationmode="country names",
        hover_name="Country name",
        size="Population",
        title=f"Global Population in {year}",
        projection="natural earth"
    )

    fig.update_layout(title_font_size=20, title_x=0.5)
    fig.show()

plot_population_map(population_data, 2020)
```

```python
import plotly.express as px

def enhanced_population_map(frame, year):

    filtered_data = frame[frame['Year'] == year]

    fig = px.scatter_geo(
        filtered_data,
        locations="Country name",
        locationmode="country names",
        hover_name="Country name",
        size="Population",
        size_max=60,
        color="Population",
        color_continuous_scale=px.colors.sequential.Rainbow,
        title=f"Population by country in {year}",
        projection="natural earth"
    )

    fig.update_traces(marker=dict(line=dict(width=0)))
    fig.update_layout(title_font_size=20, title_x=0.5)
    fig.show()

enhanced_population_map(population_data, 2000)
```

```python
import pandas as pd
import plotly.express as px

def plot_by_age_group(
    frame,
    country,
    export=False,
    export_format='html',
    export_path='population_by_age_group'
):
    """
    Plots stacked bar chart of population by age group over time for a given country.

    Parameters:
    - frame (pd.DataFrame): Full population dataset.
    - country (str): Country name to filter by.
    - export (bool): Whether to export the figure (as HTML or PNG).
    - export_format (str): 'html' or 'png'.
    - export_path (str): Path or filename for export.
    """
    # Validate inputs
    if country not in frame['Country name'].unique():
        raise ValueError(f"'{country}' not found in the dataset.")

    required_columns = [
        'Year',
        'Country name',
        'Population',
        'Population of children under the age of 15',
        'Population aged 15 to 64 years'
    ]

    for col in required_columns:
        if col not in frame.columns:
            raise KeyError(f"Missing column: {col}")

    # Filter and calculate
    country_data = frame[frame['Country name'] == country].copy()

    country_data['Pop under 15'] = country_data['Population of children under the age of 15']
    country_data['Pop 15 to 64'] = country_data['Population aged 15 to 64 years']
    country_data['Pop over 64'] = country_data['Population'] - (
        country_data['Pop under 15'] + country_data['Pop 15 to 64']
    )

    # Reshape for plotting
    age_group_data = pd.DataFrame({
        'Year': country_data['Year'],
        'Pop under 15': country_data['Pop under 15'],
        'Pop 15 to 64': country_data['Pop 15 to 64'],
        'Pop over 64': country_data['Pop over 64']
    })

    melted_data = age_group_data.melt(id_vars='Year',
                                      var_name='Age Group',
                                      value_name='Population')

    # Sort by Year
    melted_data.sort_values(by='Year', inplace=True)

    # Plot
    fig = px.bar(
        melted_data,
        x='Year',
        y='Population',
        color='Age Group',
        title=f"Population by Age Group: {country}",
        labels={'Population': 'Population (in billions)', 'Year': 'Year'},
        template='plotly'
    )
    fig.update_layout(
        barmode='stack',
        title_font_size=22,
        title_x=0.5,
        yaxis_tickformat=',',
        legend_title='Age Group',
        margin=dict(l=40, r=40, t=60, b=40)
    )

    # Show plot
    fig.show()
plot_by_age_group(population_data, 'United States')
```
